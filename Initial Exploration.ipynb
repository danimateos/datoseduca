```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import urllib
%matplotlib inline
```

```python
# B1 annex of PISA 2012
urllib.urlretrieve("http://dx.doi.org/10.1787/888932935667", "888932935667.xls")
```

```python
df = pd.read_excel("888932935667.xls", sheetname="Table I.2.27", header=13, na_values=['m'], skip_footer=6)
plt.scatter( df['GDP per capita (in equivalent USD converted using PPPs)1'], df['Mean performance on the mathematics scale'])
```

```python
num_graphs = len(df.columns[:8]) - 1


f, ax = plt.subplots(2, 4, figsize=(20,10))

for ix, col in enumerate(df.columns[1:8]):
    
    
    x = ix / 4
    y = ix % 4
    
    axis = ax[x, y]
    axis.scatter(df[col], df[df.columns[0]])
    axis.set_title(col)
    
```
